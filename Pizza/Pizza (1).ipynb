```python
import seaborn as sns
from sklearn import linear_model
import pandas as pd
```

```python
df=pd.read_csv("pizza.csv")
```

```python
df
```

```python
sns.lmplot(x='diameter',y='price',data=df)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['diameter']],df['price'])
```

```python
reg.predict([[20]])
```

```python
reg.predict([[50]])
```
