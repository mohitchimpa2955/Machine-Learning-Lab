```python
import pandas as pd
```

```python
data=pd.read_csv("training.csv")
```

```python
data
```

```python
x=data.iloc[:, :-1].values
```

```python
y=data.iloc[:, -1].values
```

```python
for i in range(len(y)):
    if y[i].strip().lower()=='yes':
        s=x[i].copy()
        break
```

```python
G=[['?' for _ in range(len(s))]]
for i in range(len(x)):
    if y[i].strip().lower()=='yes':
        for j in range(len(s)):
            if x[i][j]!=s[j]:
                s[j]='?'
        G=[g for g in G if all(g[k]=='?' or g[k]==s[k] for k in range(len(s)))]
    else:
        new_G = []
        for g in G:
            for j in range(len(s)):
                if g[i]=='?' and s[j]!=x[i][j]:
                    temp = g.copy()
                    temp[j] = s[j]
                    if temp not in new_G:
                        new_G.append(temp)
        if new_G:
            G = new_G

print("Final S:", s)
print("Final G:", G)
```
